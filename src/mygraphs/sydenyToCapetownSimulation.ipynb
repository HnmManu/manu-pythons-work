```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

sydney = (-33.8688, 151.2093)
capetown = (-33.9249, 18.4241)

plt.figure(figsize=(12,8))
m = Basemap(projection='mill', llcrnrlat=-70, urcrnrlat=10, llcrnrlon=0, urcrnrlon=180, resolution='l')

m.drawcoastlines()
m.drawcountries()
m.fillcontinents(color='lightgray', lake_color='aqua')
m.drawmapboundary(fill_color='aqua')
m.drawparallels(range(-70,11,10), labels=[1,0,0,0])
m.drawmeridians(range(0,181,20), labels=[0,0,0,1])

m.drawgreatcircle(sydney[1],sydney[0], capetown[1], capetown[0], color='blue', linewidth=2, label='Sydeny -> Capetown')

#mark cities
x1, y1 = m(sydney[1], sydney[0])
x2, y2 = m(capetown[1], capetown[0])
plt.plot(x1, y1, 'bo',markersize=6, label='Sydney')
plt.plot(x2, y2, 'go',markersize=6, label='Capetown')

plt.legend(loc='lower left')
plt.title('Simulated Flight Path: Sydney to Capetown')
plt.show()
```
