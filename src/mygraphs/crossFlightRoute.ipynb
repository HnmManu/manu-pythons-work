```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.basemap import Basemap


bengaluru = (12.9716, 77.5946)
london = (51.5074, -0.1278)
capetown = (-33.9249, 18.4241)
moscow = (55.7558, 37.6173)

plt.figure(figsize=(12,8))
m = Basemap(projection='ortho',lat_0=10,lon_0=30, resolution='l')

m.drawcoastlines()
m.drawcountries()
m.fillcontinents(color='lightgray', lake_color='aqua')
m.drawmapboundary(fill_color='aqua')
m.drawparallels(range(-90, 91, 30))
m.drawmeridians(range(-180, 181, 60))

x1, y1 = m(bengaluru[1], bengaluru[0])
x2, y2 = m(london[1], london[0])
x3, y3 = m(capetown[1], capetown[0])
x4, y4 = m(moscow[1], moscow[0])

m.drawgreatcircle(bengaluru[1], bengaluru[0], london[1], london[0], color='blue')
m.drawgreatcircle(capetown[1], capetown[0], moscow[1], moscow[0], color='green')

plt.plot(x1, y1, 'bo', markersize=6)
plt.plot(x2, y2, 'bo', markersize=6)
plt.plot(x3, y3, 'go', markersize=6)
plt.plot(x4, y4, 'go', markersize=6)

#plt.legend(loc='lower left')
plt.title("Great-circle Routes and closest Approach of two planes")
plt.show()
```
