```python
import matplotlib.pyplot as plt
import numpy as np

# Constants
latitude_degree_km = 111.32  # Distance covered by 1° latitude in km (rough average)

# Create an array of latitudes from -90° to +90°
latitudes = np.linspace(-90, 90, 1000)

# Calculate the distance covered by 1° longitude at each latitude
longitude_degree_km = latitude_degree_km * np.cos(np.radians(latitudes))

# Calculate the area of 1° x 1° cell at each latitude
area_km2 = latitude_degree_km * longitude_degree_km

# Plotting the area vs latitude
plt.figure(figsize=(10, 6))
plt.plot(latitudes, area_km2, label="Area of 1°x1° Cell", color="blue")
plt.xlabel("Latitude (degrees)")
plt.ylabel("Area (km²)")
plt.title("Area of 1°x1° Latitude-Longitude Cell vs Latitude")
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linestyle='--', linewidth=0.5, label="Equator")
plt.legend()
plt.tight_layout()
plt.show()
```
